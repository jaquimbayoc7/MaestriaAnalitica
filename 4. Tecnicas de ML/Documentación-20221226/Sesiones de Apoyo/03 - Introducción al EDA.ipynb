## EDA 


EDA (Exploratory Data Analysis) es un proceso sobre los datos para descubrir patrones, detectar anomalías, probar hipótesis y comprobar supuestos. Aunque es una tarea que puede ser transversal, es imprescindible previo a la tarea de modelación. 

El EDA se utiliza principalmente para:
- Ver lo que los datos pueden revelar.
- Proporciona una mejor comprensión de las variables del conjunto de datos y de las relaciones entre ellas.
- Puede ayudar a determinar si las técnicas de modelación o el método de análisis que estás considerando es adecuado.
- Ayuda a definir los pasos de limpieza, transformación y selección de datos.
- Confirma o definine los procesos de modelación.
- Entender si es posible alcanzar los objetivos propuestos.
- Descubrir errores o limitaciones de los datos.
- Clarificar y delimitar el problema (¿Qué pregunta intentas resolver?).
- Ayuda a responder: ¿Cómo puedes añadir, transformar, seleccionar o eliminar features para sacar más provecho de tus datos?, entre otros aspectos.

El proceso de EDA incluye:

¿Qué tipo de datos tienes y cómo tratar los diferentes tipos que encuentras?
  + Tamaños: Columnas (atributos) y filas (registros, observaciones o instancias)
  + Tipos de datos a nivel informático (tipos de variables, ...)
  + Tipos de datos a nivel estadístico (medias, medianas, cuantiles, correlaciones, ...)
  
¿Qué valores ves?, ¿Qué distribuciones encuentras? ¿Qué relaciones hay? ¿Qué patrones identificas? ¿Qué relaciones aportan o afectan a la solución del problema?
  + Análisis estadístico desccriptivo: univariado, bivariado, multivariado 
  + Análisis estadístico inferencial 
  
¿Qué errores hay en los datos y cómo se aborda esta situación?<br/>
  + Hay información duplicada, errores de captura, typos, información erronea
  
¿Qué es lo que falta en los datos y cómo se aborda esta situación?<br/>
  + Missing values
  
¿Hay valores atípicos?¿Cómo tratarlos? (No siempre se identifican en un análisis univariado, en ocasiones es necesario hacer un análsis conjunto o multivariado?
  
  
## Algunas Reflexiones con el proceso EDA:</b> <ul>
 - EL EDA no es un proceso con un conjunto de reglas y tareas estrictas a realizar. Depende del problema y de su contexto.
 -  No siempre se dispone de información sobre el dataset como lo proporcionan los ejemplos de clase, cursos o de la Web.
 - En este proceso, es necesario mantener una comunicación fluida y continua con los expertos del tema. Esto no sólo puede ahorrarte tiempo y te da respuestas, también influye en las preguntas que hagas sobre los datos.
 - En ocasiones el experto no tiene suficiente conocimiento de los datos que tienen y del problema que desea solucionar. El EDA puede ayudar a clarificar y delimitar el problema, identificar otros problemas y establecer los alcances reales.
 - No se trata de crear un informe, sino de comprender el problema.
 - Estás buscando establecer una serie de hipótesis que deseas explorar al modelar. Buscas tener una mayor comprension e intuición de los datos.
 - **Hay que pasar tiempo con los datos**
 - Evitar el sesgo de complejidad (tendencia a preferir las soluciones complejas a las simples). No siempre lo complejo y más sofisticado es lo mejor
 - Los modelos son tan buenos como las preguntas y la comprensión que tengas de los datos y del problema.
 - [GIGO](https://es.wikipedia.org/wiki/Garbage_in,_garbage_out): Garbage in, garbage out.

### Intuición sobre los tipos en los dataframes

Cuando usamos `read_csv` en pandas (y en otras librerías similares) si no especificamos los tipos de las variables **pandas intentará deducir qué tipo tiene cada variable en base al contenido**. Por ejemplo: si una columna sólo contiene números le asignará el tipo `int64`.

Esto tiene sus limitaciones:

- Es más lenta la lectura (además de leerlo, tenemos que deducir los tipos y luego convertir los datos a estos tipos).
- En algunos datasets podemos tener erratas o problemas que hagan deducir erróneamente a pandas cada tipo. Por ejemplo que las primeras filas tengan números y las últimas letras...

Por ello lo recomendable es especificar el [parámetro dtype](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) al leer CSV (especialmente los ficheros muy grandes). Por ejemplo:

```python
pd.read_csv("my.csv", dtype = {"edad": "int", "sexo": "category"})
```

De esta manera el proceso de lectura será más rápido, consumirá menos memoria y lo más importante no tendremos errores en los tipos de datos.

Esto implica que podemos tener tipos incorrectos. Por eso siempre debemos revisar `df.info()` y corregir los tipos erróneos.

#### Diferencia entre object y category

Las variables categóricas en los CSV se suelen representar con textos ('hombre', 'logroño', etc). Si se detecta texto típicamente se asigna el tipo `object` cuando se lee un CSV de forma automática.

Vamos a entender mejor `object`:

Una serie (columna de un DataFrame) de tipo `object` contiene **cualquier objeto de python en su interior**, incluído cadenas de texto (`str`). Por ejemplo:

````python
pd.Series(["alex", "marta", "carlos"])
````

Pero insisto: **cualquier tipo de objeto**

````python
pd.Series(["alex", "marta", "carlos", 5, {"a":1}])
````

Es decir, `object` no siempre es equivalente a cadenas de texto (aunque es lo más común).

Este problema (la ambigüedad de su contenido) junto al hecho de que almacenar y procesar texto es muy poco eficiente hicieron que los desarrolladores de `pandas` crearan el tipo `category`.

Intuitivamente este tipo es como un `factor` de R o un `enum` de otros lenguajes.

Lo que hará pandas es almacenar en un diccionario de equivalencias las distintas categorías y usar números (mucho más compactos y eficientes) para guardar estas variables.

Si tuvieramos una categoría de "hombre" y "mujer" podríamos asignar el 0 a hombre y el 1 a mujer y almacenar la variable como 0, 1, 1, 0, 0 según correspondiera.

Vamos a convertir una columna de tipo `object` en tipo `category`:

````python
columna_object = pd.Series(["hombre", "mujer", "hombre"])
columna_object
````

````python
columna_category = columna_object.astype("category")
````

````python
columna_category
````

Podéis ver ahora que es de tipo categórico y que hay dos categorías:

````python
columna_category.cat.categories
````

Y que internamente hombre = 0 y mujer = 1. Podemos verlo con `cat.codes`:

````python
columna_category.cat.codes
````

Por tanto ahora tendremos un `int8` (número entero) que ocupa mucho menos, es más rápido y además nos evitamos errores humanos como puede ser usar categorías incorrectas.

Las variables categóricas tienen muchas opciones como hacerlas ordinales, añadir categorías, renombrarlas, etc... Podéis ver más sobre su uso en la [guía oficial de este tipo](https://pandas.pydata.org/docs/user_guide/categorical.html).

#### Corrigiendo los tipos de las variables

Si detectamos que hay variables que debieran ser categóricas podemos:

1. Usar `dtype` en el `read_csv` desde el principio para especificar los tipos adecuados (mejor opción)
2. Usar `astype` para convertir las columnas después (más lento pero más cómodo)

````python
df.embarked = df.embarked.astype("category")
````

````python
# Podemos ver que embarked ahora es categórica:
df.info()
````

Un truco muy común es convertir todos los objetos en categorías. **CUIDADO AL HACER ESTO** ya que no siempre los `object` son variables categóricas. Aún así vamos a realizarlo y entender cuando no hay que hacerlo.

Para convertir todas usaremos `select_dtypes`:

````python
# Vamos a crear por comodidad una lista de todas las variables de tipo object:
columnas_object = df.select_dtypes(["object"]).columns.to_list()
columnas_object
````

````python
# Convertimos todas esas variables a category de golpe:
df[columnas_object] = df[columnas_object].astype("category")
````

````python
df.info()
````

Como podéis comprobar ya no quedan objects. Ya podríamos continuar con este dataset.

#### Cuándo usar category y cuándo no

¿Entonces por qué hay que tener cuidado con este método automático? Porque no siempre los object deben ser categorías. Veamos un ejemplo distinto. En esta ocasión un dataset de pokemons:

````python
df_pokemon = pd.read_csv("https://raw.githubusercontent.com/double-thinker/datasets/main/pokemon.csv")
df_pokemon.head()
````

````python
df_pokemon.info()
````

Como veis tenemos tres variables `object`: `Name`, `Type 1` y `Type 2` ¿Cuál de ellas deberíamos convertir en categorías?

Como regla general vamos a convertir a categóricas aquellas variables que representen una categoría, que tengan pocos valores respecto al dataset y que no sean identificadores únicos.

Por ejemplo: `Type 1` y `Type 2` son buenas candidatas. Sólo hay unos pocos tipos, se repiten mucho y no son identificadores únicos. Pero por otro lado `Name` no lo es: cada poemon tiene un nombre distinto, no se repiten y es un valor distinto para cada uno.

Otro buen ejemplo de variable `object` que **no** debería convertirse en categoría es el nombre de una persona o su documento de identidad.

Como veis no hay una regla totalmente clara. Pero no es recomendable convertir a categorias variables que no se repiten mucho o que no se repiten nada (son únicas).

En este caso deberíamos convertir las categóricas seleccionandolas a mano:

````python
# Método 1: en el momento de leer el csv cambiar los tipos
df_pokemon = pd.read_csv("https://raw.githubusercontent.com/double-thinker/datasets/main/pokemon.csv",
                         dtype = {"Type 1": "category", "Type 2": "category"})
df_pokemon.info()
````

````python
# Método 2: convertirlo posteriormente con listas de variables
df_pokemon = pd.read_csv("https://raw.githubusercontent.com/double-thinker/datasets/main/pokemon.csv")

variables_categoricas = ["Type 1", "Type 2"]

df_pokemon[variables_categoricas] = df_pokemon[variables_categoricas].astype("category")
````

#### Categorías con números

El formato en el que nos proporcionan los datos no siempre es el que queremos. Y muchas veces tendremos variables categóricas como números. Por ejemplo en el dataset del titanic **survived se considera numérica (usa 0s y 1s) pero a efectos de análisis deberíamos considerarla categórica**. No estamos expresando una cantidad sino una pertenencia (superviviente/ no superviviente)

````python
df.head()
````

Por ello, el método automático también falla en estas ocasiones. **Mi recomendación es escribir a mano explicitamente cuales son las variables categóricas**. Esto ayuda a corregir errores y a hacer nuestros EDAs (y modelos posteriores) más robustos (método 2 de la anterior sección).

Vamos a convertir survived y pclass:

````python
df.survived = df.survived.astype("category")
df.pclass = df.pclass.astype("category")
````

`inplace` es un parámetro que está en muchas funciones de pandas y realiza la operación **inplace** o "en el lugar". Es decir, en el mismo dataframe.

Es recomendable usar inplace cuando se pueda porque en ocasiones mejora el rendimiento.

# Valores perdidos

Una parte muy relevante del EDA es analizar los valores perdidos o NA o NaN. Son valores que no aparecen en nuestro dataset porque no se han podido recoger, no son fiables o en general no están disponibles.

Podemos contar los valores perdidos para ver la importancia en nuestro dataset:

````python
# El total de filas es 891. Todas las varibles con un conteo menor es porque tienen valores perdidos
df.count()
````

````python
df.isna().sum()
````

Podemos ver que la variable que mayor cantidad de valores perdidos tiene es `age`.

### Breve introducción a las estrategias para gestionar valores perdidos

Podemos eliminarlos con `dropna` que elimina cualquier fila con al menos un valor perdido. Esto es un riesgo, especialmente si los valores perdidos no son aleatorios.

**CUIDADO**. Los valores perdidos son aleatorios cuando no tienen ninguna relación con las variables del estudio. Es decir, han ocurrido al azar en esas posiciones como podrían haber ocurrido en otras. Pero si hay algún tipo de sesgo en estos valores perdidos usar `dropna` puede estropear todos los modelos y análisis posteriores.

Por ejemplo: si en una encuesta a familias con alumnos escolarizados en primaria les preguntamos los ingresos anuales puede que algunas familias no contesten para no reconocer su situación precaria. Es decir, **el valor perdido no es aleatorio** es más probable tener valores perdidos en familias con bajos niveles socioeconómicos. 

Si eliminara las filas con NA significaría que en mis análisis estaría sesgando la muestra y no sería representativa de las familias con menos recursos.

````python
df.dropna()
````

Podemos ver que quedan 712 filas

````python
df.drop(columns=["age"]).dropna()
````

Si quitamos la columna age antes de el dropna conservaremos más registros. Esto es útil si la variable que tiene muchos valores perdidos no se usa o no es tan relevante.

Una estrategia más avanzada es **imputar**. Hay muchas maneras pero en general es sustituir los valores perdidos con un valor razonable. Por ejemplo la media:

````python
edad_media = df.age.mean()
edad_media
````

Usaremos fillna para ello:

````python
df.age = df.age.fillna(edad_media)
````

````python
df.count()
````

Por último, no todos los valores perdidos son inadecuados. En ocasiones tienen un significado. Por ejemplo en `embarked`. El valor perdido el embarked significa que el usuario nunca embarcó al titanic. Es decir, de cara al modelo para predecir su fallecimiento esto no debe ser ignorado sino que se puede usar como un valor "especial" que es relevante.

Hay muchos modelos que pueden usar los valores perdidos como los random forest, otros no (como los modelos lineales). Pero en ocasiones merece la pena asignar un valor especial a los valores perdidos si consideramos que son relevantes y que tienen un "significado" más allá de no existir el dato

````python
df.loc[df.embarked.isna()]
````

Podemos ver que aquellas dos personas que no subieron al titanic sobrevivieron (obviamente)

Esto es útil para un modelo así que en vez de un dropna las conservaremos:

````python
df.embarked = df.embarked.cat.add_categories("No embarked")
````

````python
df.embarked = df.embarked.fillna("No embarked")
````

````python
df.count()
````
